#### yield的3种用法（生成器的3种应用模式）

- 生成器的3种模式
    - pull：特点在于能不断向外产出数据，也就是迭代器
    - push：特点在于能不断向内发送数据，比如计算移动平均值的例子，是非常早期的协程概念
    - task：任务式（是AsyncIO里的协程）

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def show_flow(path):
    img = mpimg.imread(path)
    plt.figure(figsize=(12, 8))
    plt.imshow(img, interpolation='none')
    plt.axis('off')
    plt.show()

show_flow('figs/flow_gen_coroutine.png')
```
